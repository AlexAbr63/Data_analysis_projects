## Приоритизация гипотез

### Приоритизации гипотез с помощью фреймворка ICE

```python
# Определение ICE
hypothesis['ice'] = ((hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts']).round(1)
display(hypothesis[['Hypothesis', 'ice']].sort_values(by='ice', ascending=False))
```

- Гипотезы 8 и 0 имеют самую высокую приоритизацию;
- гипотезы 7, 6 и 2 имеют среднюю приоритизацию;
- гипотезы 1, 5, 3 и 4 имеют самую низкую приоритизациюа.

### Приоритизации гипотез с помощью фреймворка RICE

```python
# Определение RICE
hypothesis['rice'] = (hypothesis['Reach'] * hypothesis['Impact'] * hypothesis['Confidence']) / hypothesis['Efforts']
display(hypothesis[['Hypothesis', 'rice']].sort_values(by='rice', ascending=False))
```

- Гипотеза 7 имеет самую высокую приоритизацию;
- гипотезы 2, 0 и 6 имеют среднюю приоритизацию;
- гипотезы 8, 3, 1, 5 и 4 имеют самую низкую приоритизацию.

## Анализ A/B-теста

### Подготовка данных

**Описание данных orders**

В датасете orders содержатся данные о совершённых заказах:

- transactionId — идентификатор заказа;
- visitorId — идентификатор пользователя, совершившего заказ;
- date — дата, когда был совершён заказ;
- revenue — выручка заказа;
- group — группа A/B-теста, в которую попал заказ.

```python
# Изучение данных
orders = orders.rename(columns={'transactionId':'transaction_id', 'visitorId':'visitor_id'})
display(orders)
print(orders.info())
```

**Описание данных visitors**

В датасете visitors содержатся данные о пользователях попавших в одну из групп теста:

- date — дата;
- group — группа A/B-теста;
- visitors — количество пользователей в указанную дату в указанной группе A/B-теста.

```python
# Изучение данных
display(visitors)
print(visitors.info())
```

```python
# Проверка на наличие дубликатов
print('Дубликатов в orders:', orders.duplicated().sum())
print('Дубликатов в visitors:', visitors.duplicated().sum())
print()

# Проверка на наличие пропусков
print('Пропусков в orders:')
print(orders.isna().sum())
print()
print('Пропусков в visitors:')
print(visitors.isna().sum())
```

```python
# Изменение типа даты
orders['date'] = orders['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
visitors['date'] = visitors['date'].map(lambda x: dt.datetime.strptime(x, '%Y-%m-%d'))
```

```python
# Проверка интервала времени
print(orders['date'].describe())
print()
print(visitors['date'].describe())
```

В результатах теста могли оказаться пользователи, которые попали в обе группы. Если такие пользователи имеются, удалю их из orders, что бы данные групп были независимыми.

```python
# Проверка, есть ли в датасете orders_A пользователи из orders_B
users_two = np.intersect1d(orders.query('group == "A"')['visitor_id'].unique(),
                               orders.query('group == "B"')['visitor_id'].unique())
print('Количество пользователей, присутствующих в обоих группах:', len(users_two))
```

```python
# Удаление пользователей, попавших в обе группы
orders = orders[np.logical_not(orders['visitor_id'].isin(users_two))]
orders
```

```python
# Количество покупателей в каждой группе
count_A = orders[orders['group'] == 'A']['visitor_id'].nunique()
count_B = orders[orders['group'] == 'B']['visitor_id'].nunique()
print('Количество покупателей в группе А:', count_A)
print('Количество покупателей в группе В:', count_B)
print('В группе В на', count_B - count_A, 'покупателя больше')
```

```python
# Количество пользователей в каждой группе
visitors_count_A = visitors[visitors['group'] == 'A']['visitors'].sum()
visitors_count_B = visitors[visitors['group'] == 'B']['visitors'].sum()
print('Количество пользователей в группе А:', visitors_count_A)
print('Количество пользователей в группе В:', visitors_count_B)
print('В группе B на', visitors_count_B - visitors_count_A, 'пользователей больше')
print('В группе B на', (100 - visitors_count_A / visitors_count_B * 100).round(2), '% пользователей больше')
```

**Вывод**

Дубликатов в датафреймах нет. Также отсутствуют пропущенные значения. Тип даты изменён на datetime гггг-мм-дд. Интервал временни исследования совпадает в обоих датасетах и составляет 2019-08-01 - 2019-08-31.

В данных обнаружены 58 пользователей, которые попали в обе группы. Такие пользователи были удалены, что бы группы остались независимыми.

Количество покупателей в группе А составляет 445, а в группе В на 83 больше - 528.

Количество посетителей в группе А составляет 18736, а в группе В на 0,95% больше - 18916.

### График кумулятивной выручки по группам

```python
# Массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()
datesGroups
```

```python
# Расчёт кумулятивных метрик заказов по дням
ordersAggregated = (datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
    .agg({'date':'max', 'group':'max', 'revenue':'sum', 'transaction_id': pd.Series.nunique, 'visitor_id': pd.Series.nunique}), axis=1).sort_values(by=['date', 'group']))

# Расчёт кумулятивных метрик визитов по дням
visitorsAggregated = (datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
    .agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1).sort_values(by=['date', 'group']))

# Создание общей таблицы
cumulativeData = ordersAggregated.merge(visitorsAggregated, on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'revenue', 'orders', 'buyers', 'visitors']
display(cumulativeData)
```

```python
# Разделение по группам для построения графиков
cumulativeRevenueA = cumulativeData[cumulativeData['group'] == 'A']
cumulativeRevenueB = cumulativeData[cumulativeData['group'] == 'B']

plt.figure(figsize=(13,7))
plt.title('Сumulative revenue')
plt.ylabel('Revenue')
plt.xlabel('Date')
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], label = 'A')
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], label = 'B')
plt.grid(True)
plt.legend()
plt.annotate('Скачок', xy=(cumulativeRevenueA.loc[32, 'date'], 3000000), xytext=(cumulativeRevenueA.loc[18, 'date'], 3500000),
            arrowprops=dict(facecolor='g'))
plt.annotate('Отрыв', xy=(cumulativeRevenueA.loc[8, 'date'], 800000), xytext=(cumulativeRevenueA.loc[8, 'date'], 2000000),
            arrowprops=dict(facecolor='g'))
plt.show()
```

**Выручка равномерно увеличивается на протяжении всего периода. На графике группы В виден большой скачок, говорящий либо о большом количестве заказов, либо о крупном заказе. ### Статистическая значимость различий в конверсии между группами по «сырым» данным

```python
# Определю количество заказов по пользователям совершивших 1 и более заказов
ordersByUsersA = (orders[orders['group'] == 'A'].groupby(['visitor_id'], as_index=False)
                                                .agg({'transaction_id':pd.Series.nunique}))
ordersByUsersA.columns = ['visitor_id', 'orders']

ordersByUsersB = (orders[orders['group'] == 'B'].groupby(['visitor_id'], as_index=False)
                                                .agg({'transaction_id':pd.Series.nunique}))
ordersByUsersB.columns = ['visitor_id', 'orders']
display(ordersByUsersA.head(3))
display(ordersByUsersB.head(3))
```

```python
# Определю количество посетителей в каждой группе по дням
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']

visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

display(visitorsADaily.head(3))
display(visitorsBDaily.head(3))
```

```python
# Создам переменные, в которых для каждого пользователя указано количество заказов, включая 0 заказов
sampleA = pd.concat([ordersByUsersA['orders'], pd.Series(0, index=np.arange(visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])), name='orders')],axis=0)
sampleB = pd.concat([ordersByUsersB['orders'], pd.Series(0, index=np.arange(visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])), name='orders')],axis=0)

display(sampleA.head(3), sampleA.tail(3))
display(sampleB.head(3), sampleB.tail(3))
sampleA
```

**Формулировка гипотез**

**Нулевая гипотеза:** Конверсии между группами А и В по «сырым» данным равны.

**Альтернативная гипотеза:** Конверсии групп А и В по «сырым» данным различаются.

```python
alpha = .05

results_1 = round((st.mannwhitneyu(sampleA, sampleB, alternative = 'two-sided')[1]), 3)
relative_change_1 = round(((sampleB.mean() / sampleA.mean() - 1) * 100))

if results_1 < alpha:
    print('p-значение:', results_1, '\nОтвергаем нулевую гипотезу о том, что статистически значимых различий в конверсии между группами нет')
else:
    print('p-значение:', results_1, '\nНе отвергаем нулевую гипотезу о том, что статистически значимых различий в конверсии между группами нет')

print('Относительное изменение группы В к группе А:', relative_change_1, '%')
```

**По результатам теста видно, что различие в конверсии групп по "сырым" данным статистически значимо. Конверсия группы В на 16% выше конверсии группы А. Но из анализа аномалий, выяснилось, что больше 2 заказов это аномалии. Они влияют на результаты теста.**

### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным.

**Формулировка гипотез**

**Нулевая гипотеза:** Средние чеки заказов между группами А и В по «сырым» данным равны.

**Альтернативная гипотеза:** Средние чеки заказов групп А и В по «сырым» данным различаются.

```python
results_2 = (st.mannwhitneyu(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'], alternative = 'two-sided')[1]).round(3)
relative_change_2 = round(((orders[orders['group'] == 'B']['revenue'].mean() / orders[orders['group'] == 'A']['revenue'].mean() - 1) * 100))

if results_2 < alpha:
    print('p-значение:', results_2, '\nОтвергаем нулевую гипотезу о том, что статистически значимых различий в среднем чеке между группами нет')
else:
    print('p-значение:', results_2, '\nНе отвергаем нулевую гипотезу о том, что статистически значимых различий в среднем чеке между группами нет')

print('Относительное изменение группы В к группе А:', relative_change_2, '%')
```

```python
# Создам переменные, в которых для каждого пользователя указано количество заказов, включая 0 заказов по отфильтрованным данным
sampleAFiltered = pd.concat([ordersByUsersA[np.logical_not(ordersByUsersA['visitor_id'].isin(abnormalUsers))]['orders'],
                             pd.Series(0,
                                       index=np.arange(visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA[np.logical_not(ordersByUsersA['visitor_id'].isin(abnormalUsers))]['orders'])),
                                       name='orders')],
                             axis=0)

sampleBFiltered = pd.concat([ordersByUsersB[np.logical_not(ordersByUsersB['visitor_id'].isin(abnormalUsers))]['orders'],
                             pd.Series(0,
                                       index=np.arange(visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB[np.logical_not(ordersByUsersB['visitor_id'].isin(abnormalUsers))]['orders'])),
                                       name='orders')],
                             axis=0)
```

**Нулевая гипотеза:** Конверсии между группами А и В по «очищенным» данным равны.

**По результатам теста видно, что различие в конверсии групп по "очищенным" данным статистически значимо. Конверсия группы В на 19% выше конверсии группы А.**

**Сравню с результатами тест по "сырым" данным:**

- Отвергаем нулевую гипотезу о том, что статистически значимых различий в конверсии между группами нет
- Относительное изменение группы В к группе А: 16 %

### Кумулятивные метрики и относительные изменения по очищенным данным
Для полноценной оценки, "какая группа лучше?", определю найденные раннее метрики, но по очищенным данным

```python
# Перезапишу переменную orders исключив из неё аномальных пользователей и использую раннее написанный код.
orders = orders[np.logical_not(orders['visitor_id'].isin(abnormalUsers))]

# Массив уникальных пар значений дат и групп теста
datesGroups = orders[['date','group']].drop_duplicates()

# Расчёт кумулятивных метрик заказов по дням
ordersAggregated = (datesGroups.apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
    .agg({'date':'max', 'group':'max', 'revenue':'sum', 'transaction_id': pd.Series.nunique, 'visitor_id': pd.Series.nunique}), axis=1).sort_values(by=['date', 'group']))

# Расчёт кумулятивных метрик визитов по дням
visitorsAggregated = (datesGroups.apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
    .agg({'date':'max', 'group':'max', 'visitors':'sum'}), axis=1).sort_values(by=['date', 'group']))

# Создание общей таблицы
cumulativeData = ordersAggregated.merge(visitorsAggregated, on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'revenue', 'orders', 'buyers', 'visitors']

# Разделение по группам для построения графиков
cumulativeRevenueA = cumulativeData[cumulativeData['group'] == 'A']
cumulativeRevenueB = cumulativeData[cumulativeData['group'] == 'B']
```

Как видно из графика, построенного по очищенным данным, отрыв в кумулятивной выручке стал более явным, а аномальный скачок больше не искажает динамику графика.

```python
# График относительного изменения кумулятивного среднего чека группы B к группе A по очищенным данным
# Объединение таблиц
mergedCumulativeRevenue = cumulativeRevenueA.merge(cumulativeRevenueB, on='date', suffixes=['A', 'B'])

# Построение графика
plt.figure(figsize=(13,7))
plt.title('Relative change in cumulative average check')
plt.xlabel('Date')
plt.plot(mergedCumulativeRevenue['date'],
         ((mergedCumulativeRevenue['revenueB']/mergedCumulativeRevenue['ordersB']) /
         (mergedCumulativeRevenue['revenueA']/mergedCumulativeRevenue['ordersA']) -1))
plt.grid(True)
plt.axhline(y=0, color='red', linestyle='--')
plt.show()
```
