#### Retention Rate
Определю Retention Rate, применяя когорный анализ и средний Retention Rate на второй месяц «жизни» когорт. Построю график изменения метрики во времени.

```python
# Расчет первого месяца активности для формирования когорт
first_activity_date = visits_log.groupby('uid')['start_ts'].min()
first_activity_date.name = 'first_activity_date'

# Слияние датасетов
visits_log = visits_log.join(first_activity_date, on='uid')

# Выделение месяца
visits_log['first_activity_month'] = visits_log['first_activity_date'].astype('datetime64[M]')

# Выделение lifetime когорт
visits_log['cohort_lifetime'] = ((visits_log['start_ts'] - visits_log['first_activity_date']) / np.timedelta64(1, 'M')).round().astype(int)

# Определение количества активных пользователей
cohorts = visits_log.groupby(['first_activity_month', 'cohort_lifetime']).agg({'uid':'nunique'}).reset_index()

# Когортный анализ по количеству посетителей
cohorts_visitors = visits_log.pivot_table(index='first_activity_month',
                                          columns='cohort_lifetime',
                                          values='uid',
                                          aggfunc='nunique')
cohorts_visitors.fillna('')
```

```python
# Определение Retention Rate
# Начальное число пользователей в когорте
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_activity_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'count_cohort_users'})

# Объединение начального числа пользователей в когорте с общими данными по когортам
cohorts = cohorts.merge(initial_users_count, on='first_activity_month')

# Определение Retention Rate
cohorts['retention'] = cohorts['uid'] / cohorts['count_cohort_users']

# Усечение даты до нужной точности
cohorts['first_activity_month'] = cohorts['first_activity_month'].dt.strftime('%Y-%m')

# Построение сводной таблицы для Retention Rate
retention_pivot = cohorts.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime',
    values='retention',
    aggfunc='sum')

# Удаление 0-го периода
retention_pivot = retention_pivot.drop(columns=0)

retention_pivot
```

```python
# Создание тепловой карты
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, vmax=0.1, fmt='.1%', linewidths=1, linecolor='gray');
```

```python
# Среднее значение Retention Rate по всем когортам на 2-ой месяц "жизни"
retention_2 = round((retention_pivot[1].mean() * 100), 1)
print('Среднее значение Retention Rate по всем когортам на 2 месяц "жизни":', retention_2, '%')
```

```python
# Среднее по lifetime когорт
retention_all = (retention_pivot.mean(axis=0) *100)

# Изменение среднего Retention Rate во времени по lifetime
retention_all[1:].plot.bar(grid=True,rot=0)
plt.title('Retention Rate by cohort')
plt.ylabel('value Retention Rate, %')
plt.show()
```

**Выводы**

В первый же месяц жизни значение Retention Rate падает до 5-7% но затем падение идёт плавное и медленное.

Среднее значение Retention Rate по всем когортам на 2 месяц "жизни": 5.5 %

По диаграмме видно, что среднее значение метрики падает с 1 когорты от 5,5% по 12 когорту до 1,3%.

Можно сделать вывод, что после первой активности на первый месяц жизни возвращается всего лишь 5-7% от начального числа новых пользователей. Но их дальнейшее поведение показывает стабильный интерес к продукту каждый месяц, хоть и плавно убывающий.

Разные когорты имеют не одинаковое количество пользователей на нулевой месяц жизни. Так, когорты, начиная с 2017-10 по 2018-03, имеют в 1,5 - 2 раза больше пользователей, чем остальные. Но, соотнося эти периоды с Retention Rate видно, что пользователи в этих когортах, всё равно продолжают возвращаться всё меньше.
Более ранние когорты имеют меньше посетителей на 0-ом месяце, в районе 10-15 тысяч человек, но их Retation Rate значительно выше. Это говорит о том, что источники, из которых пользователи приходили в период 2017-06 - 2017-09 более эффективны, ведь их Retation Rate выше.

### Метрики электронной коммерции

#### Время до покупки
Найду, сколько времени в среднем проходит с момента первого посещения сайта до совершения покупки.

```python
# Группировка данных по пользователю и определение даты первого действия
first_visit_date = visits_log.groupby('uid')['start_ts'].min().reset_index()
first_purchase_date = orders_log.groupby('uid')['buy_ts'].min().reset_index()

# Слияние датасетов
time_visit_purchase = first_purchase_date.merge(first_visit_date, on='uid', how='left')
# Исключение строк, где дата покупки меньше даты первого посещения
time_visit_purchase = time_visit_purchase[time_visit_purchase['start_ts'] <= time_visit_purchase['buy_ts']]
```

```python
# Определение времени между первым посещением и первой покупкой в секундах
delta = (time_visit_purchase['buy_ts'] - time_visit_purchase['start_ts']).dt.seconds.reset_index()
delta = delta.drop(columns=['index'])
delta.columns = ['seconds']

# Просмотр распределения данных
delta.hist(bins=100, figsize=(10, 5))
plt.show()
delta.boxplot(column='seconds', figsize=(10, 10))
plt.show()
```

Значения времени между первым посещением и первой покупкой имеют ненормальное распределение. Поэтому в качестве средней меры использую МОДУ.

```python
print('Среднее время между первым посещением и первой покупкой:', delta['seconds'].mode()[0], 'секунд')
```

#### Среднее количество покупок на одного покупателя за период
Определю среднее количество покупок на одного покупателя за 6 месяцев.

```python
orders_per_buyer = orders_log.copy()

# Добавление даты первой покупки
orders_fm = orders_per_buyer.groupby('uid')['buy_ts'].min().reset_index()
orders_fm.columns = ['uid', 'first_buy_date']
orders_per_buyer = orders_per_buyer.merge(orders_fm, on='uid', how='left')
orders_per_buyer['first_order_month'] = orders_per_buyer['first_buy_date'].astype('datetime64[M]')

# Добавление времени жизни когорты
orders_per_buyer['life_time'] = ((orders_per_buyer['buy_ts'] - orders_per_buyer['first_buy_date']) / 
                                 np.timedelta64(1, 'M')).round().astype(int)

# Определение количества покупок в каждой когорте по времени жизни
number_purchases = orders_per_buyer.groupby(['first_order_month', 'life_time']).agg({'revenue':'count'}).reset_index()
number_purchases.columns = ['first_order_month', 'life_time', 'count_revenue']

# Определение количества покупателей в каждой когорте
number_buyers = orders_per_buyer.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
number_buyers.columns = ['first_order_month', 'uid_nunique']

# Объединение таблиц и расчёт числа покупок на одного покупателя
orders_per_buyer_2 = number_purchases.merge(number_buyers, on='first_order_month')
orders_per_buyer_2['purchases_per_buyer'] = orders_per_buyer_2['count_revenue'] / orders_per_buyer_2['uid_nunique']

# Построение сводной таблицы с накопительной суммой
ddfg = orders_per_buyer_2.pivot_table(index='first_order_month',
                                      columns='life_time',
                                      values='purchases_per_buyer').cumsum(axis=1).round(2)

print('Среднее число покупок на одного покупателя по когортам за 6 месяцев жизни:', round(ddfg[6].mean(), 2))
ddfg.fillna('')
```

**Выводы**

Для когорт, которые прожили не меньше 6 месяцев, среднее число покупок на одного покупателя по когортам за 6 месяцев жизни: 1.47 покупок. К таким когортам относятся когорты с 2017-06-01 по 2017-12-01.

Значит из всех людей, которые совершают покупки, каждый делает полторы покупки в течение 6 месяцев.
