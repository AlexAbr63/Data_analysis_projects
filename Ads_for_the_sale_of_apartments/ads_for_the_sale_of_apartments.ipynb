**Пропуски в столбце floors_total**

В столбце floors_total находится 86 пропусков. При этом в этих же строках, в столбце floor имеется значение. Предполагаю, что эти пропуски вызваны тем, что квартира находится на последнем этаже дома и пользователь, указывая этаж квартиры, не стал указывать этот же этаж, как максимальный в доме. Поэтому заменяю пропуски значениями столбца floor.

```python
# Замена пропусков
df['floors_total'] = df['floors_total'].fillna(df['floor'])
print('Количество пропусков в столбце floors_total -', df['floors_total'].isna().sum())
```

**Пропуски в столбце ceiling_height**

В столбце ceiling_height имеется большое количество пропусков. Буду заполнять их медианным значением высоты потолков по домам в зависимости от максимального этажа. Т.к. высота потолков зачастую связана с количеством этажей в доме. Пропуск вызван тем, что пользователь выложивший объявление, не стал заполнять этот параметр.

```python
# Удаление строк с пропущенными значениями в столбце ceiling_height
df_not_nan = df.dropna(subset = ['ceiling_height'])
# Подсчёт медианного значения высоты потолков по домам с одинаковым количеством этажей
df_median_ceiling_height = df_not_nan.pivot_table(index='floors_total', values='ceiling_height', aggfunc='median')
```

```python
# Слияние исходного df и нового датасета с медианными значения высоты потолков
df = df.merge(df_median_ceiling_height, on='floors_total', how='left')
# Замена пропущенных значений на медианные значения
df['ceiling_height_x'] = df['ceiling_height_x'].fillna(df['ceiling_height_y'])
```

В строке номер 11575  остался пропуск поскольку квартир в 33 этажном доме в списке одна. Заполню пропуск высоты потолков вручную значением квартиры в 34 этажном доме.

```python
df['floors_total'].value_counts()
display(df[df['floors_total'] == 34])
df['ceiling_height_x'] = df['ceiling_height_x'].fillna(2.65)
print('Количество пропусков в столбце ceiling_height_x -', df['ceiling_height_x'].isna().sum())
```

**Пропуски в столбце living_area**

Поскольку жилая площадь больше всего зависит от количества комнат, буду заменять пропуски медианным значением площади по количеству комнат в квартире. Пропуск вызван тем, что человек выложивший объявление, не стал заполнять этот параметр.

```python
# Удаление строк с пропущенными значениями в столбце living_area
df_not_nan_living_area = df.dropna(subset = ['living_area'])
# Подсчёт медианного значения жилой площади для каждого количества комнат
df_median_living_area = df_not_nan_living_area.pivot_table(index='rooms', values='living_area', aggfunc='median')
# Слияние исходного df и нового датасета с медианными значениями жилой площади
df = df.merge(df_median_living_area, on = 'rooms', how = 'left')
# Замена пропущенных значений на медианные значения
df['living_area_x'] = df['living_area_x'].fillna(df['living_area_y'])
print('Количество пропусков в столбце living_area_x -', df['living_area_x'].isna().sum())
```

**Пропуски в столбце balcony**

В данном столбце 11 519 пропусков. Через команду df['balcony'].value_counts() определил, что количество балконов изменяется от 0 до 5. Вероятно, пропуск вызван тем, что в квартире балкона нет, и этот параметр заполнять не стали. Заполню пропуски значением - 0.

```python
df['balcony'] = df['balcony'].fillna(0)
print('Количество пропусков в столбце balcony -', df['balcony'].isna().sum())
```

**Пропуски в столбце kitchen_area**

Т.к. площадь кухни относится к общей площади квартиры, пропуски в этом столбце буду заполнять медианным значением kitchen_area по total_area. Для того что бы посчитать медиану, категоризую данные общей площади. Что бы понять на какие интервалы разбить значения оценю их с помощью гистограммы. Для этого строю гистограмму по столбцу total_area. Для категоризации данных выбираю значения 25, 50, 75, 100, 125, 150+. Пропуск вызван тем, что пользователь выложивший объявление, не стал заполнять этот параметр.

```python
# Построение гистограммы
df['total_area'].plot(kind = 'hist',
                 y = 'total_area',
                 range = (15, 300),
                 bins = 20,
                 grid = True,
                 legend = True)
plt.title('histogram of total_area 2015-2019')
plt.xlabel('total_area', fontsize = 12, color = 'Black')
plt.ylabel('Frequency \n шт.')

# Создание функция для разбиения на категории
def category_total_area (s):
    if s <=25:
        return 'до 25'
    if s <=50:
        return 'до 50'
    if s <=75:
        return 'до 75'
    if s <=100:
        return 'до 100'
    if s <=125:
        return 'до 125'
    if s <=150:
        return 'до 150'
    if s > 150:
        return '150+'
# Создание нового столбца и заполнение его в соответсвие с функцией
df['total_area_category'] = df['total_area'].apply(category_total_area)
# Очистка в отдельном датафремйе столбца kitchen_area
df_not_nan_kitchen_area = df.dropna(subset = ['kitchen_area'])
# Создание сводной таблицы с медианными значениями площади кухни для разных значений общей площади
df_median_kitchen_area = df_not_nan_kitchen_area.pivot_table(index = 'total_area_category',
                                                             values = 'kitchen_area',
                                                             aggfunc = 'median')
# Слияние исходного df и нового датасета с медианными значениями площади кухни
df = df.merge(df_median_kitchen_area, on = 'total_area_category', how = 'left')
# Замена пропущенных значений на медианные значения
df['kitchen_area_x'] = df['kitchen_area_x'].fillna(df['kitchen_area_y'])
```

```python
display(df.sample(3))
print('Количество пропусков в столбце kitchen_area_x -', df['kitchen_area_x'].isna().sum())
```

Итак, квартиры на промежуточных этажах дороже в отличие от первого и последнего этажей.

```python
# Функция для перевода категорий этажей к формату 1, 2 и 3
def floor_column_last_price (h):
    if h['floor_column'] == 'первый':
        return 1
    if h['floor_column'] == 'другой':
        return 2
    if h['floor_column'] == 'последний':
        return 3
# Добавление нового столбца
df['floor_1_2_3'] = df.apply(floor_column_last_price, axis = 1)
# Построение диаграммы рассеяния
df.plot(x = 'floor_1_2_3',
        y = 'last_price',
        kind = 'scatter',
        alpha = 0.1)
plt.title('dependence of the last_price on floor')
plt.ylabel('last_price \n млн.р.')
plt.xlabel('floor \n этаж')
plt.show()
# Расчёт коэф.кор.
print(df['last_price'].corr(df['floor_1_2_3']))

# Построение диаграмм ящик с усами
df.groupby('floor_column').boxplot(column = 'last_price', figsize = (5, 10))
plt.show()
```

Зависимость дня недели, месяца и года от последней цены.

Коэф.кор. зависимостей около нуля. Это говорит об отсутствии связи между этими параметрами. Построение диаграмм это подтверждает. Медианное значение цены квартир практически не меняется ни с изменением дня недели, ни с изменением месяца, ни с изменением года.

```python
# Построение диаграмм ящик с усами для дней недели
df.groupby('weekday').boxplot(column = 'last_price', figsize = (5, 12))
plt.show()

# Построение диаграмм ящик с усами для месяцев
df.groupby('month').boxplot(column = 'last_price', figsize = (5, 16))
plt.show()

# Построение диаграмм ящик с усами для годов
df.groupby('year').boxplot(column = 'last_price', figsize = (5, 13))
plt.show()
```

Определю топ 10 населённых пунктов по числу объявлений и найду для них среднюю цену квадратного метра.

```python
# Группирую города и считаю количество публикаций
df_count = df.groupby('locality_name').count().sort_values(by = 'last_price', ascending = False)
# Нахожу топ 10
display(df_count.iloc[:10])
# Формирую новый датасет с топ 10 городов по количеству публикаций 
df_10 = df.query('locality_name in ("санкт-петербург", "посёлок мурино", "посёлок шушары", "всеволожск", "колпино", "посёлок парголово", "деревня кудрово", "пушкин", "гатчина", "выборг")')
# Подсчёт средней цены квадратного метра
df_price_square_meter = df_10.pivot_table(index = 'locality_name', values = 'price_square_meter', aggfunc =['mean'])
# Переименование столбцов
df_price_square_meter.set_axis(['mean_price_square_meter'], axis = 'columns', inplace = True)
# Приведение данных к формату тысячи рублей и сортировка
df_price_square_meter['mean_price_square_meter'] = (df_price_square_meter['mean_price_square_meter'] / 1000).round().astype(int)
df_price_square_meter.sort_values(by = 'mean_price_square_meter', ascending = False)
```
