## Предобработка данных

```python
# Приведение названий столбцов к нижнему регистру
df.set_axis(['name', 'platform', 'year_of_release', 'genre', 'na_sales', 'eu_sales',
       'jp_sales', 'other_sales', 'critic_score', 'user_score', 'rating'], axis = 'columns', inplace = True)
```

```python
# Приведение слов в столбцах к нижнему регистру
df['name'] = df['name'].str.lower()
df['platform'] = df['platform'].str.lower()
df['genre'] = df['genre'].str.lower()
```

```python
# Перевод значений в столбце year_of_release из типа float к int
df['year_of_release'] = df['year_of_release'].astype('Int64')
```

```python
# Проверка на наличие дубликатов
df.duplicated().sum()
```

В столбце df['user_score'] имеется 2424 строчки с значением tbd вместо значения оценки. Это "To Be Determined", т.е. "предстоит определить" и оценки ещё нет. Поскольку продажи по этим строчкам имеются, значит, игра всё-таки вышла. Предположу, что значения отсутствуют из-за того, что информация собиралась из разных источников и в некоторых, значение рейтинга не обновляли. Заменю аббревиатуру пропуском, и изменю тип данных с object на float.

```python
# Замена tbd на NaN
df.loc[df['user_score'] == 'tbd', 'user_score'] = np.nan
# Изменеие типа данных
df['user_score'] = df['user_score'].astype(float)
```

```python
# Поиск пропусков
print(df.isna().sum())
```

# Пропуски в столбце name
В столбце присутствуют 2 пропуска. Они появились в результате ошибки при выгрузке или отсутствовали в первоисточнике. Их удалять не буду, поскольку в этой строчке есть информация о платформе, о годе и о количестве копий, а всего строчек с платформой gen - 29. При таком малом количестве 2 строчки важны. Заменю пропуски на unnamed.

```python
# Строчки с пропусками
display(df[df['name'].isna()])
# Количество строк с платформой gen
print((df['platform'] == 'gen').sum())
# Заполнение пропусков
df['name'] = df['name'].fillna('unnamed')
```

# Пропуски в столбце genre
Пропуски в этом столбце находятся в тех же строчках, что и пропуски в столбце name (и по тем же причинам). Поскольку узнать жанр не представляется возможным, заменю пропуски на unknown genre.

```python
# Строчки с пропусками
display(df[df['genre'].isna()])
# Заполнение пропусков
df['genre'] = df['genre'].fillna('unknown genre')
```

# Пропуски в столбце year_of_release
Пропуски вызваны тем, что игру добавили в списки до анонса и в последствие значение не заполняли. Значения можно было бы заполнить по тем же играм, но на другой платформе, где дата стоит, но тогда возможно внесение погрешности в данные. Эти пропуски заполнять не буду.

# Пропуски в столбце rating
В данном столбце пропуски ничем заполнить не получится т.к. нет данных для их корректного заполнения. Пропуски вызваны тем, что информация в первоисточнике не заполнялась.

# Пропуски в столбце critic_score и user_score
Пропущенные значения рейтингов заполнять не буду, поскольку нет объективной информации, по которой это можно было бы сделать. Рейтинг не зависит ни от жанра, ни от платформы, ни от года выпуска и любая игра могла, как провалиться, так и стать успешной. Единственный показатель, с которым можно связать рейтинг это продажи. Ведь чем игра успешнее, тем больше продажи. Но заполнять в связке с продажами так же нет смысла, поскольку для ответа на вопрос исследования можно использовать сами же продажи, а не восстановленные по ним рейтинги. Пропуски вызваны тем, что информация в первоисточнике не заполнялась.

```python
# Преобразование к целочисленному типу
df['critic_score'] = df['critic_score'].astype('Int64')
```

```python
# Подсчёт суммарных продаж во всех регионах
df['total_sales'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

 - Перевёл значения в столбце year_of_release из типа float к int
 - Проверил отсутствие дубликатов
 - В столбцах name и genre заполнил пропуски
 - Пропуски в столбцах year_of_release, rating, critic_score и user_score не заполнял, что бы не исказить данные

## Проверка гипотез

### Пользовательские рейтинги платформ Xbox One и PC
Сформулирую нулевую гипотезу и противоположную ей альтернативную:
- Н0 - Средние пользовательские рейтинги платформ Xbox One и PC равны.
- Н1 - Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Сначала визуализирую распределение каждой выборки, чтобы определить нормальное ли оно или нет.

```python
# Построение гистограмм
fig_5, axes = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 7))

axes.hist(df_2010_pc['user_score'])
axes.hist(df_2010_xone['user_score'])

axes.set_xlabel('Estimation')
axes.set_ylabel('Frequency')
axes.set_title('Histogram of user ratings')
axes.grid()
```

Как видно из гистограмм обе представлены распределением немного скошенным влево.

Проверю дисперсии выборок, чтобы корректно определить атрибут equal_val в Т–тесте.

```python
print('Дисперсия для XONE', round(np.var(df_2010_xone['user_score']), 2))
print('Дисперсия для PC', round(np.var(df_2010_pc['user_score']), 2))
```

Т.к. дисперсии примерно одинаковые, параметр equal_val считаю True.

```python
# Для корректной работы Т-теста избавляюсь от пропусков в столбцах
df_2010_xone = df_2010_xone.dropna(subset = ['user_score'])
df_2010_pc = df_2010_pc.dropna(subset = ['user_score'])
```

```python
# Проверка гипотезы
alpha = 0.05 # критический уровень статистической значимости принимаю равным 5%

results = st.ttest_ind(df_2010_xone['user_score'], df_2010_pc['user_score'])

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```
