В этом датафрейме нет пропусков и полных дубликатов. Но есть повторяющиеся product_id с разными ценами. Разберу их подробнее. Все сильно отличающиеся product_id отсортировались вначале списка.

```python
# Сортировка шифров product_id
product_codes['product_id'].value_counts().reset_index().sort_values(by='index', ascending=False).head(30)
```

```python
# Функция для выборочного изучения шифров product_id в обоих датафреймах
def info_product_id(id):
    display(product_codes[product_codes['product_id'] == id])
    display(retail_dataset[retail_dataset['product_id'] == id])

info_product_id('AMAZONFEE')
```

gift_0001_50, gift_0001_40, gift_0001_30, gift_0001_20 - это подарки выданные магазином. У них есть стоимость, но количество в таблице retail_dataset указано 0. Поэтому при расчёте суммы при умножении на ноль получится ноль. Оставлю их.

POST вероятно - это товары отправленные почтой или товары, заказанные в интернет магазине и полученные в конкретном магазине. Они имеют разную стоимость. Поскольку нельзя установить конкретных связей возьму среднее значение их стоимости.

BANK CHARGES - это банковские сборы. Они имеют разные цены за единицу, а в таблице retail_dataset либо нулевое, либо отрицательное количество. Это значит, что в общую сумму они должны попасть либо с минусом и вычитаться из общего дохода, либо не попасть вовсе. Возьму среднее значение.

AMAZONFEE - судя по информации из интернета это комиссия за продажи на AMAZON. Ситуация такая же как с BANK CHARGES. Возьму среднее значение.

Понять значения остальных шифров нет возможности, но судя по расшифрованным, ошибок в данных нет, а необходимые действия учитываются в таблице retail_dataset. Для правильного объединения таблиц, позиции в product_codes должны быть в единичном экземпляре. Т.к. значения сильно отличаются в пределах одного product_id уместно взять медианное значение. Взять значение, которое использовалось последним не подходит, т.к., например, в течение 2 месяцев значение могло быть маленьким, а в последний стать большим или стать большим и не продаваться, это внесёт ложную информацию в данные. Поэтому в качестве средней меры использую медиану.

```python
# Создам новый датафрейм с единичными product_id и усреднёнными ценами
product_codes_new = (product_codes.groupby('product_id', as_index = False)['price_per_one']
                                  .median()
                                  .sort_values(by='product_id', ascending = False))
product_codes_new
```

```python
# Подсчёт товаров с ценой 0
print('Количество товаров с ценой за единицу равной 0:', len(product_codes_new[product_codes_new['price_per_one']==0]), 'шт.')
```

После расчёта медианной стоимости для каждого товара, осталось 57 позиций с ценой ноль. Вероятно, это позиции, которые внесли в базу, но не заполнили стоимость. Удалю такие позиции.

```python
# Удаление строк с нулевыми ценами
product_codes_new = product_codes_new[product_codes_new['price_per_one'] != 0]
```

#### Слияние датафреймов

#### Пропуски

Попробую заменить пропуски в стобце buyer_id на значения id которые содержатся в одном чеке с пропуском. Для этого создам датафрейм, в котором каждому значению receipt_id соответствует одно значение buyer_id.

```python
# Создание датафрема
receipt_buyer = retail[['receipt_id', 'buyer_id']].dropna().drop_duplicates()
receipt_buyer['receipt_id'].value_counts()
```

При подсчёте и сортировке видно, что все значения повторяются один раз (т.к. отсортированы). Объединю с итоговым датафреймом и посмотрю, сколько пропусков заменилось.

```python
retail.merge(receipt_buyer, on='receipt_id', how='left').info()
```

Количество пропущенных значений осталось прежним, а значит восстановить id покупателя нет возможности. Заменю пропуски на значение unknown, и оставлю эти строки, т.к. информация по ним имеется и может пригодиться.

```python
# Замена пропусков на unknown
retail['buyer_id'] = retail['buyer_id'].fillna('unknown')
retail.info()
```

### Расчёт показателей по клиентам

#### Среднее значение количества товаров в чеке по неделям по группам клиентов программы лояльности
Поскольку для нахождения этого параметра необходимо использовать id покупателей, необходимо исключить данные с неизвестным id, т.к. все строки с неизвестным id будут считаться одним покупателем.

```python
# Добавление столбца с неделями
retail['week'] = retail['purchase_date'].astype('datetime64[W]')

# Датасет без клиентов с неизвестным id
retail_not_unknown = retail[retail['buyer_id'] != 'unknown']

# Определение среднего количества товаров для участников программы лояльности и обычных клиентов
retail_1 = retail_not_unknown[retail_not_unknown['loyalty_program'] == 1]
retail_0 = retail_not_unknown[retail_not_unknown['loyalty_program'] == 0]

mean_quantity_products_1 = (retail_1.groupby('receipt_id').agg({'quantity':'sum', 'week':'min'})
                                    .groupby('week', as_index = False)['quantity'].median())

mean_quantity_products_0 = (retail_0.groupby('receipt_id').agg({'quantity':'sum', 'week':'min'})
                                    .groupby('week', as_index = False)['quantity'].median())

mean_quantity_products_1['loyalty_program'] = 1
mean_quantity_products_0['loyalty_program'] = 0

# Слияние датасетов
mean_quantity_products = pd.concat([mean_quantity_products_1, mean_quantity_products_0], axis=0, ignore_index=True)
```

```python
# Построение графика
plt.figure(figsize=(14, 5))
plt.title('График среднего количества товаров в чеке')
sns.lineplot(x='week', y='quantity', data = mean_quantity_products, hue = 'loyalty_program', legend=False)
plt.xlabel('Неделя')
plt.ylabel('Количество товаров в чеке')
plt.legend(['0','1'], title = 'Участники программы лояльности')
plt.show()
```

`
Графики среднего количества товаров в чеке имеют схожую тенденцию и несколько раз пересекаются, за исключением предновогоднего периода. С 15 декабря наблюдается резкий рост значений, но у клиентов программы лояльности пик значительно выше.
Клиенты покупают примерно одинаковое количество товаров, кроме 2 недель перед новым годом, когда клиенты участники программы лояльности покупают значительно больше товаров.
`

#### Среднее количество покупок (чеков) по неделям по группам клиентов программы лояльности на одного клиента (частота покупок)

Поскольку для нахождения этого параметра используется id покупателей, необходимо исключить данные с неизвестным id, т.к. все строки с неизвестным id будут считаться одним покупателем.

```python
# Создание датасета чек-неделя-программа лояльности
retail_not_unknown = retail[retail['buyer_id'] != 'unknown']
mean_quantity_receipt = retail_not_unknown.groupby('receipt_id', as_index = False).agg({'week':'min', 'loyalty_program':'mean'})

# Определение количества чеков по неделям по группам программы лояльности
mean_quantity_receipt_1 = mean_quantity_receipt[mean_quantity_receipt['loyalty_program'] == 1].groupby('week', as_index = False)['receipt_id'].count()
mean_quantity_receipt_0 = mean_quantity_receipt[mean_quantity_receipt['loyalty_program'] == 0].groupby('week', as_index = False)['receipt_id'].count()

# Определение количества уникальных пользователей по неделям по группам
number_unique_buyers_week_1 = retail_not_unknown[retail_not_unknown['loyalty_program'] == 1].groupby('week', as_index = False).agg({'buyer_id':'nunique'})
number_unique_buyers_week_0 = retail_not_unknown[retail_not_unknown['loyalty_program'] == 0].groupby('week', as_index = False).agg({'buyer_id':'nunique'})

# Слияние датасетов
mean_quantity_receipt_buyers_1 = mean_quantity_receipt_1.merge(number_unique_buyers_week_1, on = 'week')
mean_quantity_receipt_buyers_0 = mean_quantity_receipt_0.merge(number_unique_buyers_week_0, on = 'week')

# Определение среднего количества покупок
mean_quantity_receipt_buyers_1['mean'] = (mean_quantity_receipt_buyers_1['receipt_id'] / mean_quantity_receipt_buyers_1['buyer_id']).round(2)
mean_quantity_receipt_buyers_1['loyalty_program'] = 1

mean_quantity_receipt_buyers_0['mean'] = (mean_quantity_receipt_buyers_0['receipt_id'] / mean_quantity_receipt_buyers_0['buyer_id']).round(2)
mean_quantity_receipt_buyers_0['loyalty_program'] = 0

mean_quantity_receipt_buyers = pd.concat([mean_quantity_receipt_buyers_1, mean_quantity_receipt_buyers_0], axis=0, ignore_index=True)
mean_quantity_receipt_buyers_ = pd.concat([mean_quantity_receipt_buyers_1, mean_quantity_receipt_buyers_0], axis=1)
mean_quantity_receipt_buyers_.columns = ['week', 'receipt_id_1', 'buyer_id_1', 'mean_1', 'loyalty_program_1', 'week_0', 'receipt_id_0', 'buyer_id_0', 'mean_0', 'loyalty_program_0']
mean_quantity_receipt_buyers_
```

```python
# Построение графика
plt.figure(figsize=(13, 5))
plt.title('График среднего количества покупок на клиента')
sns.lineplot(x='week', y='mean', data = mean_quantity_receipt_buyers, hue = 'loyalty_program', legend=False)
plt.xlabel('Неделя')
plt.ylabel('Количество покупок на клиента')
plt.legend(['0','1'], title = 'Участники программы лояльности')
plt.show()

# Построение графика относительного изменения среднего количества покупок на клиента обычных клиентов к клиентам программы лояльности
plt.figure(figsize=(9,3))
plt.title('График относительного изменения среднего количества покупок клиентов 0 к 1 ')
plt.xlabel('Week')
plt.plot(mean_quantity_receipt_buyers_['week'],
         ((mean_quantity_receipt_buyers_['mean_0']) /
         (mean_quantity_receipt_buyers_['mean_1']) -1))
plt.grid(True)
plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=0.07, color='black', linestyle='--')
plt.show()
```

`
Из графика видно, что динамика покупок примерно равна, но клиенты программы лояльности совершают немного меньше покупок.
`

#### Средний доход с покупки на одного клиента

```python
# Создание столбца со стоимостью
retail['price_quantity'] = retail['quantity'] * retail['price_per_one']
retail_not_unknown['price_quantity'] = retail_not_unknown['quantity'] * retail_not_unknown['price_per_one']

# Расчёт среднего дохода с покупки на одного клиента для участников программы лояльности
price_quantity_week_1 = (retail_not_unknown[retail_not_unknown['loyalty_program'] == 1]
                         .groupby('receipt_id', as_index = False).agg({'price_quantity':'sum', 'week':'min'})
                         .groupby('week', as_index = False)['price_quantity'].median())
price_quantity_week_1 = price_quantity_week_1.merge(number_unique_buyers_week_1, on='week')
price_quantity_week_1['loyalty_program'] = 1
price_quantity_week_1['mean_of_buyers'] = (price_quantity_week_1['price_quantity'] / price_quantity_week_1['buyer_id']).round(2)

# Расчёт среднего дохода с покупки на одного клиента для не участвующих в программе лояльности 
price_quantity_week_0 = (retail_not_unknown[retail_not_unknown['loyalty_program'] == 0]
                         .groupby('receipt_id', as_index = False).agg({'price_quantity':'sum', 'week':'min'})
                         .groupby('week', as_index = False)['price_quantity'].median())
price_quantity_week_0 = price_quantity_week_0.merge(number_unique_buyers_week_0, on='week')
price_quantity_week_0['loyalty_program'] = 0
price_quantity_week_0['mean_of_buyers'] = (price_quantity_week_0['price_quantity'] / price_quantity_week_0['buyer_id']).round(2)

# Объединение датасетов
price_quantity_week = pd.concat([price_quantity_week_1, price_quantity_week_0], axis=0, ignore_index=True)

# Построение графика
plt.figure(figsize=(9, 4))
plt.title('График среднего дохода с покупки на одного клиента')
sns.lineplot(x='week', y='mean_of_buyers', data = price_quantity_week, hue = 'loyalty_program', legend=False)
plt.xlabel('Неделя')
plt.ylabel('Доход с покупки на одного клиента')
plt.legend(['0','1'], title = 'Участники программы лояльности')
plt.show()

# Относительное изменение
price_quantity_week_ = pd.concat([price_quantity_week_1, price_quantity_week_0], axis=1)
price_quantity_week_.columns = ['week', 'price_quantity_1', 'buyer_id_1', 'loyalty_program_1',
       'mean_of_buyers_1', 'week_0', 'price_quantity_0', 'buyer_id_0',
       'loyalty_program_0', 'mean_of_buyers_0']
price_quantity_week_.head(2)

# Построение графика относительного изменения среднего количества покупок на клиента обычных клиентов к клиентам программы лояльности
plt.figure(figsize=(9,3))
plt.title('График относительного изменения среднего дохода с покупки на одного клиента 0 к 1 ')
plt.xlabel('Week')
plt.plot(price_quantity_week_['week'],
         ((price_quantity_week_['mean_of_buyers_0']) /
         (price_quantity_week_['mean_of_buyers_1']) -1))
plt.grid(True)
plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=-0.5, color='black', linestyle='--')
plt.show()
```

`
Динамика графиков очень схожа. Но клиенты программы лояльности приносят больше прибыли. В период 2 недель до нового года наблюдаются пики, причём у клиентов программы лояльности пиковые значения в несколько раз больше.
`

## Проверка гипотез

### Сравнение среднего чека клиентов программы лояльности и обычных клиентов
Ключевым критерием при оценке эффективности является доходность. Что бы оценить доходность от клиентов участвующих в программе лояльности и от обычных клиентов, сравню их средние чеки.

```python
# Подготовка массивов
mean_receipt_sample_1 = (retail[retail['loyalty_program'] == 1]
                         .groupby('receipt_id', as_index=False)
                         .agg({'price_quantity':'sum'}))

mean_receipt_sample_0 = (retail[retail['loyalty_program'] == 0]
                         .groupby('receipt_id', as_index=False)
                         .agg({'price_quantity':'sum'}))

# Расчёт средних значений
print('Значение среднего чека клиентов программы лояльности:', round(mean_receipt_sample_1['price_quantity'].mean(), 2))
print('Значение среднего чека обычных клиентов:', round(mean_receipt_sample_0['price_quantity'].mean(), 2))
print('Средний чек обычных клиентов больше на:', round((mean_receipt_sample_0['price_quantity'].mean() - mean_receipt_sample_1['price_quantity'].mean()), 2), 'у.е.')

print('Размер mean_receipt_sample_1:', len(mean_receipt_sample_1['price_quantity']))
print('Размер mean_receipt_sample_0:', len(mean_receipt_sample_0['price_quantity']))

# Визуализация распределений
fig = plt.figure(figsize = (10, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.set_title('Распределение sample_1')
ax2.set_title('Распределение sample_2')

ax1.hist(mean_receipt_sample_1['price_quantity'], bins=100)
ax2.hist(mean_receipt_sample_0['price_quantity'], bins=100)

ax1.set_ylabel('Частота')
ax1.set_ylabel('Частота')

fig.tight_layout()
```

Проведу тест о равенстве средних двух выборок. Т.к. данные содержат большие положительные и отрицательные значения использую тест Манна-Уитни:

**Гипотеза Н0**: Средний чек клиентов программы лояльности и средний чек обычных клиентов равны.  
**Гипотеза Н1**: Средний чек клиентов программы лояльности не равен среднему чеку обычных клиентов.

```python
# Проверка гипотезы о равенстве средних
sample_1 = mean_receipt_sample_1['price_quantity']
sample_2 = mean_receipt_sample_0['price_quantity']

# критический уровень статистической значимости задам на уровне 5%
alpha = .05

results = st.mannwhitneyu(sample_1, sample_2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Нулевую гипотезу отвергнуть не удалось, а значит статистически значимой разницы в значении средних нет. Средние чеки клиентов программы лояльности и обычных клиентов равны.**

### Сравнение среднего чека интернет магазина и обычных магазинов

```python
# Подготовка массивов
retail.groupby(['shop_id', 'receipt_id'], as_index=False).agg({'price_quantity':'sum'})

mean_receipt_online = (retail[retail['shop_id'] == 'Shop 0'].
                       groupby(['receipt_id'], as_index=False)
                       .agg({'price_quantity':'sum'}))

mean_receipt_offline = (retail[retail['shop_id'] != 'Shop 0'].
                       groupby(['receipt_id'], as_index=False)
                       .agg({'price_quantity':'sum'}))

# Расчёт средних значений
print('Средний чек интернет магазина:', round(mean_receipt_online['price_quantity'].mean(), 2), 'у.е.')
print('Средний чек обычного магазина:', round(mean_receipt_offline['price_quantity'].mean(), 2), 'у.е.')
print('Средний чек обычного магазина больше на:', round((mean_receipt_offline['price_quantity'].mean() -
                                                   mean_receipt_online['price_quantity'].mean()), 2), 'у.е.')

print('Размер mean_receipt_online:', len(mean_receipt_online['price_quantity']))
print('Размер mean_receipt_online:', len(mean_receipt_offline['price_quantity']))

# Визуализация распределений
fig = plt.figure(figsize = (10, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

ax1.set_title('Распределение sample_1')
ax2.set_title('Распределение sample_2')

ax1.hist(mean_receipt_online['price_quantity'], bins=100)
ax2.hist(mean_receipt_offline['price_quantity'], bins=100)

ax1.set_ylabel('Частота')
ax1.set_ylabel('Частота')

fig.tight_layout()
```

Проведу тест о равенстве средних двух выборок. Т.к. данные содержат большие положительные и отрицательные значения использую тест Манна-Уитни:

**Гипотеза Н0**: Средние чеки интернет магазина и обычного магазина равны.  
**Гипотеза Н1**: Средний чек интернет магазина не равен среднему чеку обычного магазина.

```python
# Проверка гипотезы о равенстве средних
sample_1 = mean_receipt_online['price_quantity']
sample_2 = mean_receipt_offline['price_quantity']

# критический уровень статистической значимости задам на уровне 5%
alpha = .05

results = st.mannwhitneyu(sample_1, sample_2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Нулевая гипотеза отвергается, а значит разница средних чеков статистически значима. Средний чек обычного магазина больше среднего чека интернет магазина.**

Динамика продаж для лояльных и нелояльных клиентов по неделям очень схожа. Отличие заключается в количестве покупок, у нелояльных клиентов этот показатель больше примерно в 1,5 раза. Магазин Shop 0 имеет подавляющее количество покупок. При этом по всем клиентам наблюдается неделя "затишья" когда продаж очень мало, с 22 декабря до конца года. Также в декабре продаж в целом примерно в 2 раза больше, чем в январе или феврале. Т.е. клиенты покупают в преддверии нового года значительно больше, чем в последующие месяцы.

У лояльных клиентов наблюдается среднее количество товаров на уровне 150-200, с пиками на последнюю неделю года и последнюю неделю февраля. У нелояльных клиентов наблюдается снижение количества товаров за покупку до 100 под конец года (29 декабря). По нелояльным клиентам динамика по 3 магазинам примерно схожа, но у  магазина Shop 1 большее количество товаров за покупку, далее идёт магазин Shop 4, а затем Shop 0. По лояльным клиентам и нелояльным клиентам примерно сравнимы магазины Shop 0 и shop 1 соответственно.

Средняя сумма покупок для клиентов программы лояльности находится на уровне 300-400 у.е. и имеет пики до 700 в конце декабря и конце февраля. Средняя сумма покупок для остальных клиентов снижается к 29 декабря, но всё же имеет пиковое значение на конец года и также на конец февраля. Динамика клиентов программы лояльности и обычных клиентов схожа, за исключением отсутствия просадки перед концом года для клиентов программы лояльности. По значениям магазина Shop 0 видно, что клиенты программы лояльности тратить немного больше, чем обычные покупатели.  
Динамика офлайн и онлайн магазинов схожа. К середине декабря наблюдается снижение среднего чека, но затем к концу года резкий рост. Затем снова небольшое снижение. Второй пик наблюдается к концу февраля, что связано с получением зарплат, а также низкой активностью в январе. Наблюдается разница в значениях. В офлайн магазинах средний чек примерно на 20% выше.

**Общее**  
В целом заметна тенденция, что к середине декабря покупатели совершают покупки реже, покупают товаров меньше и на меньшие суммы, но в период с 15 декабря до конца года, проявляют большую активность (тратят и покупают в разы больше и чаще). Затем, в январе показатели падают до значений меньших, чем в начале декабря и удерживаются в феврале. Вероятно после больших затрат перед завершением года, люди в начале нового года тратят менее охотно. Тем не менее, в конце января и конце февраля наблюдается небольшое увеличение активности вероятно связанное с начислением заработной платы.

Также, заметна интересная особенность, что в преддверии нового года, когда увеличивается активность клиентов, по некоторым параметрам, до этого отстающие лояльные клиенты начинают резко обгонять нелояльных клиентов или же рост показателей лояльных клиентов сильно вырастает по сравнению с нелояльными, а затем в новом году всё возвращается на исходную. Возможными причинами могут являться:

    1) К концу года клиенты чаще совершают крупные покупки, и использование карты лояльности становится особенно выгодным. Возможно, обычно клиенты не каждый раз использую карту лояльности, но крупные покупки проводят по ней (заметно на графике среднего дохода с покупки на одного клиента).
    2) Так же увеличивается количество товаров в чеке, что наталкивает на мысль, что картой клиента может пользоваться не только её владелец, но и, например, друзья или родственники (заметно на графике среднего количества товаров в чеке).

Нулевая гипотеза отвергается, а значит разница средних чеков статистически значима. Средний чек обычного магазина больше среднего чека интернет магазина.
